```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(101, 500, 100)
a = np.linspace(0.011, 0.02, 100)
plt.plot(x, np.sin(x * a))
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.show()
```
